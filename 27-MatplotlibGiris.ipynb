```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#ikide bir plt.show yazmamıza gerek olmadan da grafiği gösterir
#jupyter notebooka has bir özelliktir
```

```python
yas_liste= [10,15,20,25,30,35,40,45,50,55]
kilo_liste=[30,60,80,82,84,85,83,81,80,86]
#bir kişinin hangi yaşta kaç kilo olduğuna dair liste
```

```python
numpy_yas_liste = np.array(yas_liste)
numpy_kilo_liste = np.array(kilo_liste)
#numpy dizisine çevirdik. çümkü genelde numpy ile işlemler yaparız
#aldığımız verilerle numpy üzerinden işlem yaparız sıkça
```

```python
numpy_yas_liste
```

```python
numpy_kilo_liste
```

```python
plt.plot(numpy_yas_liste,numpy_kilo_liste,"g")
#ilk verilen liste/dizi x eksenine, 2. verilen y eksenine yazıldı
#parantez içinde yazdığım g, greeni yani plottaki rengi temsil eder
```

```python
plt.plot(numpy_yas_liste,numpy_kilo_liste,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
#eksenlerin başlığını yazdırdık
# jupyter notebook yerine başka bir script editörde yazsam plt.show()
#olarak göstermesini istemem gerekirdi
```

```python
plt.plot(numpy_yas_liste,numpy_kilo_liste,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo-Yas Degisim Grafigi")
#grafiğin ama başlığını attım
```

```python
numpy_dizisi_1 = np.linspace(0,10,20)
```

```python
numpy_dizisi_1
```

```python
numpy_dizisi_2 = numpy_dizisi_1 **3
```

```python
numpy_dizisi_2
```

```python
plt.plot(numpy_dizisi_1,numpy_dizisi_2, "r")
```

```python
plt.plot(numpy_dizisi_1,numpy_dizisi_2, "r--")
#kesikli çizgi yaptı
```

```python
plt.plot(numpy_dizisi_1,numpy_dizisi_2, "r*-")
# verilerin olduğu yerlere yıldızlar koydu
```

```python
plt.subplot(1,2,1) #1 sıra, 2 kolon olaca, şu an 1. grafik çiziliyor
plt.plot(numpy_dizisi_1,numpy_dizisi_2,"r*-")
plt.subplot(1,2,2)  #1 sıra, 2 kolon olaca, şu an 2. grafik çiziliyor
plt.plot(numpy_dizisi_2,numpy_dizisi_1,"r*-")
```

```python
benim_figürüm = plt.figure()
#bir figür oluşturdum (şu an boş)
figure_eksenleri = benim_figürüm.add_axes([0.2,0.2,0.9,0.9])
#argümanlar liste içinde verilmelidir
#sırasıyla soldaki eksenin başlangıç nokt.,alt taraftaki eksenin başlangıç nokt
# width ve height
figure_eksenleri.plot(numpy_dizisi_1,numpy_dizisi_2,"g")
figure_eksenleri.set_xlabel("X Ekseni Veri İsmi")
figure_eksenleri.set_ylabel("Y Ekseni Veri İsmi")
figure_eksenleri.set_title("Grafik Başlığı")
```

```python
figür_2 = plt.figure()
eksen_1 = figür_2.add_axes([0.1,0.1,0.7,0.7])
eksen_2 = figür_2.add_axes([0.3,0.3,0.3,0.3])

eksen_1.plot(numpy_dizisi_1,numpy_dizisi_2,"g")
eksen_1.set_xlabel("X Ekseni")
eksen_1.set_ylabel("Y Ekseni")
eksen_1.set_title("Ana Grafik Başlık")


eksen_2.plot(numpy_dizisi_2,numpy_dizisi_1,"b")
eksen_2.set_xlabel("X Ekseni")
eksen_2.set_ylabel("X Ekseni")
eksen_2.set_title("Küçük Grafik Başlık")
```

```python
figür_2 = plt.figure()
eksen_1 = figür_2.add_axes([0.1,0.1,0.9,0.9])
eksen_2 = figür_2.add_axes([0.3,0.6,0.3,0.3])

eksen_1.plot(numpy_dizisi_1,numpy_dizisi_2,"g")
eksen_1.set_xlabel("X Ekseni")
eksen_1.set_ylabel("Y Ekseni")
eksen_1.set_title("Ana Grafik Başlık")


eksen_2.plot(numpy_dizisi_2,numpy_dizisi_1,"b")
eksen_2.set_xlabel("Küçük X Ekseni")
eksen_2.set_ylabel("Küçük Y Ekseni")
eksen_2.set_title("Küçük Grafik Başlık")
```

```python
(figürüm, benim_eksenler) = plt.subplots(nrows=1,ncols=2)
type(benim_eksenler)

for eksen in benim_eksenler:
    eksen.plot(numpy_dizisi_1,numpy_dizisi_2,"g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
#eksenler arası mesafeyi düzeltti
```

```python
yeni_figür = plt.figure()
yeni_eksen =yeni_figür.add_axes([0.1,0.1,0.9,0.9])
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**2,label="Numpy Dizisi **2")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**3,label="Numpy Dizisi **3")
yeni_eksen.legend() #lejant ekleyelim
```

```python
yeni_figür = plt.figure()
yeni_eksen =yeni_figür.add_axes([0.1,0.1,0.9,0.9])
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**2,label="Numpy Dizisi **2")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**3,label="Numpy Dizisi **3")
yeni_eksen.legend(loc=1) #lejantın yerini bu şekilde değişiriz
```

```python
yeni_figür = plt.figure()
yeni_eksen =yeni_figür.add_axes([0.1,0.1,0.9,0.9])
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**2,label="Numpy Dizisi **2")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**3,label="Numpy Dizisi **3")
yeni_eksen.legend(loc=2) 
```

```python
yeni_figür = plt.figure(figsize=(3,3)) #figürün büyüklüğü (width,height) tuple olarak girilebilir. herhangi bir girdi vermezsek eksenin büyüklüğüne göre oto oluşturuluyor 
yeni_eksen =yeni_figür.add_axes([0.1,0.1,0.9,0.9])
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**2,label="Numpy Dizisi **2")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1**3,label="Numpy Dizisi **3")
yeni_eksen.legend(loc=2) 
```
