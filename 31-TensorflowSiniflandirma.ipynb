```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
#30 tane nöron koy, activation relu olsun
model.add(Dense(units=15,activation="relu")) #15 nöron
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) #tek nöronlu output layerına activation olarak sigmoid verdik

model.compile(loss="binary_crossentropy",optimizer="adam")
#sınıflandırma yaptığımız için mse yerine binary_crossentropy kullandık
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1)
```

```python
#epoch sayısını bilerek yüksek (abartı) verdik
```

```python
model.history.history
```

```python
model_kaybi = pd.DataFrame(model.history.history)
```

```python
model_kaybi.plot()
#beklediğimiz gibi loss azalmış, validation loss artmış. yani overfitting olmuş
```

```python
#elimizdeki veriyi oldukça doğru bulacaktır fakat yeni verdiğimiz bir veriyi hatalı hesaplayacaktır
```

```python
#val. loss saçma derece artarsa, epochsu durduralım ki overfitting önlensin
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
#30 tane nöron koy, activation relu olsun
model.add(Dense(units=15,activation="relu")) #15 nöron
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) #tek nöronlu output layerına activation olarak sigmoid verdik

model.compile(loss="binary_crossentropy",optimizer="adam")
#sınıflandırma yaptığımız için mse yerine binary_crossentropy kullandık
```

```python
early_stopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),callbacks=[early_stopping])
```

```python
model_kaybi= pd.DataFrame(model.history.history)
```

```python
model_kaybi.plot()
#görüldüğü gibi baya bir iyileşme oldu
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(rate=0.6)) #genelde 0.5 üstü seçilmez biz öylesine seçtik
#dropout ile unitleri(nöronları) rastgele kapatıp açıyoruz
#30 tane nöron koy, activation relu olsun
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(rate=0.6))#15 nöron
model.add(Dense(units=15,activation="relu"))
model.add(Dropout(rate=0.6))
model.add(Dense(units=1,activation="sigmoid")) #tek nöronlu output layerına activation olarak sigmoid verdik

model.compile(loss="binary_crossentropy",optimizer="adam")
#sınıflandırma yaptığımız için mse yerine binary_crossentropy kullandık
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),callbacks=[early_stopping])
```

```python
kayipDf= pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
#bir nebze de olsa düzeldi
```

```python
tahminler = model.predict_classes(x_test)
#sınıfları tahmin ettireceğiz(web sitesi zararlı ise 1 değilse 0)
```

```python
tahminler
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminler))
# y test ile bizim x test ile yapılan tahminlerimizi verip kıyaslayacağız
#%90 %92 gibi bir duyarlılıkla (precision) tahmin yapmışız, gayet iyibir oran
```

```python
print(confusion_matrix(y_test,tahminler))
#6 tane değeri yanlış, kalanları ise doğru bulmuş
```
