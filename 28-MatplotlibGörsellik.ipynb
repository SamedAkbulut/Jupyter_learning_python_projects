```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpy_dizisi_1 = np.linspace(0,10,20)
```

```python
numpy_dizisi_2 = numpy_dizisi_1**2
```

```python
numpy_dizisi_1
```

```python
numpy_dizisi_2
```

```python
(benim_figur , benim_eksen) = plt.subplots()
benim_eksen.plot(numpy_dizisi_1,numpy_dizisi_2,color="#99FFCC")
benim_eksen.plot(numpy_dizisi_2,numpy_dizisi_1,color="#FF33FF", alpha =0.5) #alpha azaldıkça saydamlaşır
#aynı tabloya 2 tane bastırdık
#özel renkler seçmek için rgb hex code ile kodları internetten bulup color olarak yazmak
```

```python
(yeni_figur,yeni_eksen) = plt.subplots()
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+2,color="blue", linewidth=6)
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+3,color="green", linewidth=1)
#linewidth çizgi kalınlığı
#+ yazmak izgiyi öteledi
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+4,color="red", linestyle="--")
#linestyle çizgi stilini değiştirir
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+5,color="purple", linestyle=":")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+6,color="black", linestyle="-.")
yeni_eksen.plot(numpy_dizisi_1,numpy_dizisi_1+7,color="orange", linestyle="--" , marker="o",markersize= 7 ,markerfacecolor="black")
#marker ile işaret markersize ile işaretin boyutu markerfacecolor ile de işaretin rengini ayarladık
```
